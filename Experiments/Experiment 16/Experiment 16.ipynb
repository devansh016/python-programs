Write a Python Program for Linear Search.

```python
nums = []
n = int(input("Enter number of elements : "))
for i in range(0, n):
    element = int(input())
    nums.append(element)

num = int(input('Enter a number to search: '));
flag = 0;
for i in nums:
  if i==num:
    print("Number Found.")
    flag=1;
    break;

if flag==0:
  print("Number not Found.")
```

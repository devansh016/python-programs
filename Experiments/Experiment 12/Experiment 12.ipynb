Write a Python Program to show the usage of multiple decorators to a single function.

```python
def decor1(func):
    def inner():
        x = func()
        return x +10
    return inner
  
def decor(func):
    def inner():
        x = func()
        return x - x
    return inner
  
@decor1
@decor
def num():
    return 10
  
print(num())
```

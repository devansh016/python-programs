```python
import numpy as np
import numpy as np
x = [1, 2, 3, 4 ,5]
print("Original array:")
print(x)
print("Reverse array:")
x = x[::-1]
print(x)
```

2.   Write a NumPy program to create a 2-dimensional array of size 2 x 3 also print the shape, type and data type of the array.

```python
import numpy as np
x = np.array([[2, 4, 6], [6, 8, 10]], np.int32)
print("Type: ", type(x))	
print("Shape: ", x.shape)
print("Data Type: ", x.dtype)
```

4.   Write a NumPy program to multiply and add two 3 x 3 matrices.

```python
import numpy as np
p = [[1, 0, 2], [0, 1, 3], [2, 1, 1]]
q = [[1, 2, 3], [3, 4, 1], [3, 3, 1]]
result1 = np.dot(p, q)
result2 = np.add(p,q)
print("Multiplication:\n" ,result1)
print("Addition:\n" ,result2)
```

5.   Write a NumPy program to compute the trigonometric sine, cosine and tangent array of angles given in degrees.

```python
import numpy as np
angle = [0., 30., 45., 60., 90.]
print("sine:")
print(np.sin(angle) * np.pi / 180.)
print("cosine:")
print(np.cos(angle) * np.pi / 180.)
print("tangent:")
print(np.tan(angle) * np.pi / 180.)
```

6.   Write a NumPy program to compute e^x, element-wise of a given array.


```python
import numpy as np
x = np.array([1., 2., 3., 4.], np.float32)
print("Original array: ")
print(x)
print("e^x:")
r = np.exp(x)
print(r)
```

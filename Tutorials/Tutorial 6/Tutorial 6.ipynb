WAP to convert the elements of two lists into key-value pairs of a dictionary.

```python
listday = ["Mon", "Tue", "Wed", "Thr", "Fri", "Sat", "Sun"]
listdate = [1, 2, 3, 4, 5, 6, 7]

print("List of days: ", listday)
print("list of date : ", listdate)
# Empty dictionary
res = {}
# COnvert to dictionary
for key in listday:
   for value in listdate:
      res[key] = value
      listdate.remove(value)
      break
print("Dictionary from lists :\n ",res)
```

WAP to find the common elements in two lists.

```python
first = [1, 2, 3, 5]
second = [2, 4, 5, 7]
common = set(first).intersection(second)
print(common)
```

WAP to modify an existing element of a tuple with a new element.

```python
num = (10, 20, 30, 40, 50) 
print(num)

lst= [int(input('Enter a new element:'))] 
new = tuple(lst)

pos = int(input('Enter position no:'))
num1 = num[0:pos-1]
num1 = num1+new 

num = num1+num[pos:] 
print(num)
```

WAP to delete an element from a particular position in a tuple.

```python
a = (1, 2, 3, 4, 5)
ls_a = list(a)
pos = int(input('Enter position no:'))
del ls_a[pos]
b = tuple(ls_a)
print(b)
```

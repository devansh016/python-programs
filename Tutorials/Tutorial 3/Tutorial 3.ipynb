Pattern 1:

```python
rows = 5;
for i in range(rows+1):
    for j in range(i):
        print(i, end=' ')
    print('')
```

Pattern 2:

```python
rows = 5
for i in range(1,rows+1):
  for j in range(rows+1-i):
    print(i, end=' ')
  print('')
```

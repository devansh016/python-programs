Write a Python Program to find the sum of a Series 1/1! + 2/2! + 3/3! + 4/4! +……. + n/n!

```python
def fact(num):
  fac = 1
  for i in range(1,num+1):
    fac = fac * i;
  return fac

n = int(input('Enter value of n: '))
sum =0;
for i in range(1,n+1):
  sum = sum + n/fact(n)
print(sum)
```

Write a Python program to print all even numbers in a range.

```python
a = int(input('Enter lower limit of range: '))
b = int(input('Enter upper limit of range: '))

for i in range(a,b+1):
  if(i%2==0):
    print(i, end=' ')
```

Show that the outputs of functions Filter, Map and Reduce can be obtained using List Comprehension.







```python
my_list = [i for i in range(10)]
print(my_list)
map(lambda x: x+1, my_list)
filter(lambda x: x>2 and x<5, my_list)
```
